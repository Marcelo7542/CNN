```python
import tensorflow as ts
import pandas as pd
import numpy as np
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt
```

```python
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") / 255
print(flower)
```

```python
fig, axs = plt.subplots(1,2,  figsize=(10, 8))
axs[0].imshow(flower)
axs[1].imshow(china)
```

```python
images = np.array([flower, china])
print(images.shape)
```

```python
batch_size, h, w, channels = images.shape
```

```python
filters = np.zeros(shape = (9,9, channels, 2), dtype = np.float32)
filters[4, :, :, 0] = 1
filters[:, 4, :, 1] = 1
```

```python
print(filters)
```

```python
result = ts.nn.conv2d(images, filters, 1, "SAME")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0,0].imshow(result[0, :, :, 0])
ax[0,1].imshow(result[0, :, :, 1])
ax[1,0].imshow(result[1, :, :, 0])
ax[1,1].imshow(result[1, :, :, 0])
```
